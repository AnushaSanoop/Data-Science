```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([5,10,15,20,25])
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(1,10)
y=np.arange(1,10)
plt.plot(x,y,"o")
plt.show()
```

```python
x=np.array([1,8])
y=np.array([3,10])
plt.plot(x,y,"o",mec="g",ms=20)
plt.xlabel("distance")
plt.ylabel("time")
plt.show()
```

```python
x=np.arange(-4,4)
y=np.array([1,9,8,3,7,4,6,10])
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(-4,4)
y=np.array([1,9,8,3,7,4,6,10])
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([5,10,15,20])
y=np.array([5,10,15,20])
plt.xlabel("distance")
plt.ylabel("time")
plt.plot(x,y,mec="g",mfc=5)
plt.show()
x=np.array([4,7,10,13,16])
y=np.array([9,4,8,5,7])
plt.plot(x,y)
plt.show
```

```python
x=np.array([5,10,15,20])
y=np.array([5,10,15,20])
plt.xlabel("distance")
plt.ylabel("time")
plt.subplot(1,2,1)
plt.title("speed")
plt.plot(x,y,color="#756435")
plt.show()
x=np.array([4,7,10,13,16])
y=np.array([9,4,8,5,7])
plt.subplot(1,2,2)
plt.title("velocity")
plt.plot(x,y,color="#996485")
plt.show
```

```python
x=np.array([1,5,9,14])
y=np.array([9,6,17,4])
clrs=np.array(["#215487","#842016","#026598","#909090"])
plt.scatter(x,y,c=clrs)
plt.show()
x=np.array([4,8,6,9])
y=np.array([1,6,3,7])
plt.scatter(x,y,c=clrs)
plt.show()
```

```python
x=np.arange(0,45,0.1)
y=np.sin(x)
plt.title("sin(x) and cos(x)")
plt.plot(x,y,color="#458625",label="sin")
y=np.cos(x)
plt.plot(x,y,color="#642076",label="cos")
plt.legend()
```

```python
x=np.random.randint(100,size=3)
plt.hist(x,width=3)
```

```python
# Bar diagram

import matplotlib.pyplot as plt
import numpy as np
x=np.array([4,8,7])
plt.xlabel("distance")
plt.ylabel("time")
y=np.array([8,1,7])
plt.bar(x,y,color="#010101",width=0.4)
```

```python
dict={"Adil":500,"Amjada":540,"Anagha":510,"Anjana":489,"Anusha":450,"Ayesha":490}
a=list(dict.keys())
b=list(dict.values())
x=np.array(a)
y=np.array(b)
plt.bar(x,y,width=0.5)
```
